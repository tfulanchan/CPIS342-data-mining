```python
# installing dependencies
import pandas as pd
import numpy as np
import seaborn as sns
import pandas as pd
death = pd.read_excel("C:/Users/chanhen/Desktop/Fall2023/CPIS342/assignments/covid19_raw.xlsx", sheet_name='TotalDeath_Pool')
```

```python
import matplotlib.pyplot as plt

# Now the 'time_column' becomes the index of the DataFrame
# You can access the time data using the index
time = death.index

# Calculate the sum of data for each time
sum_data = death.sum(axis=1)

# Plot the graph
plt.figure(figsize=(10, 6))
plt.scatter(time, sum_data, s=100, c='b', alpha=0.5)  # Scatter plot with time on x-axis and sum of data on y-axis
for i, name in enumerate('Country,Other'):
    plt.text(time[i], sum_data[i], name, fontsize=8, ha='right')  # Annotate each spot with the corresponding name
plt.xlabel('Time')
plt.ylabel('Sum of Data')
plt.title('Sum of Data vs. Time for Different Entities')
plt.show()
```
